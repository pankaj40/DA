```python
from bs4 import BeautifulSoup as bs
```

```python
import requests
```

```python
import pandas as pd
```

```python
url='https://www.amazon.in/s?k=5G+MOBILES&rh=n%3A1389401031&ref=nb_sb_noss'
```

```python
data=requests.get(url)
```

```python
data
```

```python
soup=bs(data.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
det=soup.find_all('h2',class_='a-size-mini a-spacing-none a-color-base s-line-clamp-2')
```

```python
len(det)
```

```python
name=[]
for i in range(0,len(det)):
    name.append(det[i].get_text())
```

```python
len(name)
```

```python
len(name)
```

```python
price=soup.find_all('span',class_='a-price-whole')
```

```python
price
```

```python
prices=[]
for i in range(0,len(price)):
    prices.append(price[i].get_text())
```

```python
len(prices)
```

```python
prices
```

```python
rating=soup.find_all('span',class_='a-icon-alt')
```

```python
ratings=[]
for i in range(0,len(rating)):
    ratings.append(rating[i].get_text())
```

```python
len(ratings)
```

```python
len(ratings)
```

```python
len(ratings)
```
